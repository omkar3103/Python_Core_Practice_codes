1Ô∏è‚É£ Square Numbers

Create a list of numbers from 1 to 20.

```python
l_1=[]
for i in range(1,21):
  l_1.append(i ** 2)
print(l_1)
```

2Ô∏è‚É£ Even Filter

```python
l_2=[]
for i in range(1,11):
  if (i %2 == 0):
    l_2.append(i)
print('Even Number Filtered list:',l_2)
```

```python
even_l2= [i for i in range(1,21) if i%2==0 ]
even_l2
```

üü° 3Ô∏è‚É£ Remove Negatives

```python
l_3=[123,2,-44,554,-32,-2]
l_3_1=[]
for i in l_3:
  if i>0:
    l_3_1.append(i)
print('positive Filtered :',l_3_1)
```

```python
positive_l = [x for x in l_3 if x>0]
positive_l
```

üü° 4Ô∏è‚É£ Flatten Nested List

```python
l_4=[[1,2],[3,4],[5,6]]
l_4_1=[]
for i in l_4:
  l_4_1.extend(i)
print(l_4_1)
```

```python
nested_to_list=[x for sublist in l_4 for x in sublist]
nested_to_list #we need to use sublist to flatten nested list
```

6Ô∏è‚É£ Word Length Mapping

```python
words = ["AI","Machine","Learning","Deep",'AI']
word_freq = {word: words.count(word) for word in list(words)}
word_freq
```

8Ô∏è‚É£ Conditional Comprehension

```python
nums = [1,2,3,4,5,6,7,8,9]
ev_set = {x if x%2==0 else 0 for x in nums}
ev_set
```
