simple Calculator

```python
def calculator(a,b,operator):
  if(operator=='+'):
    return a+b
  elif(operator=='-'):
    return a-b
  elif(operator== '*'):
    return a*b
  elif(operator =='/'):

    return a/b
  elif(operator=='%'):
    return a%b
  else:
    return 'Invalid Operator'

print('Result According to operator :',calculator(10,20,'-'))
```

2️⃣ Even or Odd Function

```python
def is_even(num):
  if(num%2==0):
    return 'Even'
  else:
    return 'Odd'

num=int(input('Enter a number :'))
print(is_even(num))
```

3️⃣ Prime Checker Function

```python
def is_prime(num):
   if num <=1:
    return 'Not Prime'
   for i in range(2,num):
    if num %i ==0:
      return 'Not Prime'
   else:
    return 'Prime'

num=int(input('Enter a number :'))
print(is_prime(num))
```

4️⃣ Factorial Using Function

```python
def fact(num):
  i=1
  fct=1
  while(i<=num):
    fct=fct*i
    i+=1

  return fct

num=int(input('Enter a number :'))
print(fact(num))
```

5️⃣ Reverse a Number

```python
def rev_num(num):
  rev=0
  while(num>0):
    rem=num%10
    rev=rev*10+rem
    num=num//10
  return rev

num=int(input('Enter a number :'))
print(rev_num(num))
```

6️⃣ Second Largest in List

```python
list=[13,43,414,4534,44]

def second_high(list):
  max=list[0]
  second_max=0

  for i in list:
    if(i>max):
      second_max=max
      max=i
    elif(i>second_max):
        second_max=i
  return second_max

print(second_high(list))
```

7️⃣ Count Frequency

```python
list_1=[43,41,31,31,13,13,43,43,43,31]
def frequency_counter(list_1):
  return {x: list_1.count(x) for x in set(list_1)}


print(frequency_counter(list_1))
```
