##### 3Ô∏è‚É£ Prime detector (optimized)
##### Check whether a number is prime.
##### Stop checking once a divisor is found
##### Do NOT check till full range

```python
no = int(input('Enter Number :'))

if no < 1:
   print('NO is not prime')
else:
    is_prime = True
    for i in range(2,int(no ** 0.5)+1):
        if no % i == 0:
            is_prime = False
            break

    if is_prime :
        print(no, 'Is Prime ')
    else:
        print(no,'Not Prime')
```

##### Given a number, find the first repeating digit.

```python
no = input('Enter number :')
rept = set()

for i in no:
    if i in rept:
        print('First repeating no :',i)
        break
    rept.add(i)
else:
    print('no repeating Digit')
```

```python
num = input("Enter a number: ")   # Take number as text

seen = set()                      # Empty memory box

for digit in num:                 # Check each digit one by one
    if digit in seen:             # Have I seen this digit before?
        print("First repeating digit:", digit)
        break                     # Stop immediately
    seen.add(digit)               # Remember this digit
else:
    print("No repeating digit")
```

### While loop

#### WHAT IS A WHILE LOOP? (CORE IDEA)

A while loop repeats code as long as a condition is TRUE.

üëâ Unlike for, it is used when:

You don‚Äôt know how many times loop will run

Loop depends on user input / condition

You need dynamic stopping (very ML-like)

```python
n = int(input('Enter Number :'))

while(n>0):
    print('Number is :',n)
    n=n-1
```
