Q4ï¸âƒ£ Random Arrays

Create a 3Ã—3 matrix of random numbers between 0 and 1

Find:

Maximum value

Minimum value

Mean

Standard deviation

```python
rand_arr = np.random.randint(1,10,size=(3,3))
print(rand_arr)
```

```python
print(rand_arr.max())  #Maximum value
print(rand_arr.min())  #Minimum value
print(rand_arr.mean()) #Mean
print(rand_arr.std())  #Standard deviation
```

Q6ï¸âƒ£ Boolean Masking

Extract elements greater than 20

Replace elements less than 20 with 0

```python
arr = np.array([10,15,20,25,30,35])
```

```python
for i in arr:
  if i>20:
    print(i,'')
```

```python
arr[arr<20] = 0
print(arr)
```
