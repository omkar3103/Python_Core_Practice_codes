ðŸ”¹ 6. Custom Error using raise

Ask user to enter salary.

If salary < 0, raise ValueError("Salary cannot be negative")

Handle it using tryâ€“except.

```python
class SalaryError(Exception):
  pass
try:
  salary=int(input('Enter Salary :'))
  if(salary<0):
    raise SalaryError('Salary Cannot Be Negative')
  elif(str ==type(salary)):
    raise SalaryError('Enter valid Input')
except SalaryError as se:
    print(f'{se}')
```

ðŸ”¹ 7. Password Validation

Ask user to enter password.

If password length < 6, raise error.

Handle it and show proper message.

```python
class PassLength(Exception):
  pass
try:
  password=input('Enter Password :')
  if (len(password)<6):
    raise PassLength('Password is too short')

  print('password Accepted')
except PassLength as ps:
    print(ps)
```

ðŸ”¹ 8. Calculator with Exception Handling

Create a mini calculator:

Ask user for two numbers

Ask for operation (+, -, *, /)

Handle:

invalid number input

division by zero

invalid operator

```python

class  CalculatorError(Exception):
  pass
try:
  n1=int(input('Enter Number :'))
  n2=int(input('Enter Second Number:'))
  oprtor=input('Enter Operator')
  if ('+' == oprtor):
    print(n1+n2)
  elif('-'==oprtor):
    print(n1-n2)
  elif('/'==oprtor):
    print(n1/n2)
  elif('*'==oprtor):
    print(n1*n2)
  elif('%'==oprtor):
    print(n1%n2)
  else :
      raise CalculatorError('Invalid operator')
except ZeroDivisionError:
  print('Cannot Divide by Zero')
except ValueError:
  print('Enter Valid Value')
except CalculatorError as ce:
  print(ce)
```
