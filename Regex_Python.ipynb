Regex = Pattern Matching Tool

It is used to:

```python
import re
```

re.search()

```python
text = 'Hey I am Omkar Kulkarni' #find first Search match
print(re.search('Omkar',text))
```

re.findall()

```python
print(re.findall('Kulkarni',text)) #Returns all matches
```

re.match()

```python
print(re.match('Hey',text)) #Checks match only at beginning
```

re.sub()

```python
#Replace pattern
print(re.sub('Omkar Kulkarni','Omkar Manoj Kulkarni',text))
```

Assignment

**1.Extract digits**

```python
text= 'My Omkar Kulkarni, Roll No 55 and marks are 78.34'
print(re.findall(r'\d',text)) # without using +
print(re.findall(r'\d+',text))# with using +
```

**2.Extract words**

```python
text_1='Omkar Manoj Kulkarni'
print(re.findall(r'\w+',text_1))# with using +
print(re.findall(r'\w',text_1))# without using +
```

3.Validate Email (Basic Example)

```python
email = 'My name is omkar@gmail.com'
pattern=r'\w+@\w+\.com'
print(re.findall(pattern,email))
```
